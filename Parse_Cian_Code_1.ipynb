# Cian Parsing

```python

import requests
from bs4 import BeautifulSoup
import csv

""" Plan

1. Выяснить количество страниц
2. На каждой странице определить ads (объявления)
3. Для каждого обьявления на странице собрать - цена, описание, локация


"""
def get_html (url):
    r= requests.get(url)
    return (r.text)
def get_total_pages(html):
    soup=BeautifulSoup(html, 'lxml')
    
    # find-first element with defined div/class
    # find all-elements with defined div/class
    
    # найти все обьявления на странице
  #last page
    pages=soup.find('div',class_="_93444fe79c--wrapper--2B3If").find('li', class_="_93444fe79c--list-item--2KxXr _93444fe79c--list-item--active--3dOSi").text  
    total_pages=pages
    print(int(total_pages))
    return int(total_pages)

def write_csv(data):
    # a - mean we will add new data to the end of the file
    with open('cian.csv','a',encoding='utf-8') as f:
        print('csv open')
        #писатель пишет csv в кач аргумента наш файл
        writer=csv.writer(f)
        writer.writerow( (data['room_m2_floor'],
                          data['url1'],
                          data['address'],
                          data['price'] ))
        
    
# hub for results of another functions
def get_page_data(html):
    soup=BeautifulSoup(html,'lxml')
    #контейнер который содержит интерес нас обьявления - catalog-list
    #внутри этого контейнера каждое обьявл item
    ads=soup.find('div',class_="_93444fe79c--wrapper--E9jWb").find_all('div',class_="c6e8ba5398--offer-container--2sOFx")
    
    for ad in ads:
        # title, price, metro,url
        try:
            # исп try т.к. не все продукты имеют цену или метро и в этом случ ошибка
            #извлекаем текст изописания
            # strip отсекает символы слева и справа (|n)
            #
            #room_m2_floor=ad.find('div', class_="c6e8ba5398--single_title--22TGT")
            room_m2_floor=ad.find('div', class_='c6e8ba5398--title--2CW78').text

        except:
            #room_m2_floor=ad.find('div', class_='c6e8ba5398--title--2CW78').text
            room_m2_floor=' '
            
        try:
            url1=ad.find('div', class_='c6e8ba5398--info-section--Sfnx- c6e8ba5398--main-info--oWcMk').find('a', class_="c6e8ba5398--header--1fV2A").get('href')
            #print(url1)
        except:
            url1=' '
            
        try:
            price=ad.find('div', class_='c6e8ba5398--header--1df-X').text.strip()
        except:
            price=' '
        try:
            address=ad.find('div', class_='c6e8ba5398--address-links--1tfGW').text.strip()
        except:
            address=' '
            
        data={'room_m2_floor':room_m2_floor,
                  'url1':url1,
                  'address':address,
                 'price':price }
        print(data)
        write_csv(data)
        
            
def main():
# https://www.avito.ru/moskva/telefony?q=%D1%82%D0%B5%D0%BB%D0%B5%D1%84%D0%BE%D0%BD%D1%8B+HTC&p=1
# т.к. есть измен элемент, то разобьем на 2 базовый url
    url='https://kaluga.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&p=54&region=4780&room3=1'
    
    base_url='https://kaluga.cian.ru/cat.php?deal_type=sale&engine_version=2&offer_type=flat&'
    page_url='p='
    query_part='&region=4780&room3=1'
    
    
    total_pages=get_total_pages(get_html(url))
    #total_pages=3
    
    print('Done')
    #print(total_pages)
    
    for i in range(1,total_pages+1):
        print(i)
        url_gen=base_url+page_url+str(i)+query_part
        #print(url_gen)
        #Обращаемся к опред странице
        html=get_html(url_gen)
        get_page_data(html)


# entrance for function call

if __name__=='__main__':
    main()
```
